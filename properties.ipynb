# Dataset Properties
Before Starting, it would probably be beneficial to take a quick look at what exactly our database even is. I know its a 106 column, 206 MB Gargantuan CSV, but what is there to know?
> This is also basic familizarization with numpy and pandas which probably helps...

```python
# Let's First Import this thing
import numpy as np
import pandas as pd

dataframe = pd.read_csv('./history.csv')
```

```python
# This tells us the rows x columns we are dealing with. Damn!
print(dataframe.shape)
dataframe.head(10)
```

```python
# What are the columns?
for column in dataframe.columns:
    print(column)
```

```python
# Here are a few of the bigger Categorical Data Items
grouping = dataframe.groupby(['type_code', 'label']).id.agg('count')
print(grouping)
print(grouping.agg('sum'))
```

```python
# Here are a few of the bigger Categorical Data Items
grouping = dataframe.groupby(['is_registered', 'label']).id.agg('count')
print(grouping)
print(grouping.agg('sum'))
```
